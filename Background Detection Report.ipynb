### EM-алгоритм для восстановления смеси многомерных нормальных распределений.

В рамках задания был реализован EM-алгоритм для восстановления смеси многомерных нормальных распределений. Ниже приведены результаты тестирования алгоритма.

Сгенерируем данные &mdash; точки на плоскости, разделенные на 3 группы, так что точки каждой из этих групп генерируются из многомерного нормального распределения со своим центром.

```python
X_test, y_test = gen_dataset(1000, 2, 3)
X_test = sklearn.preprocessing.StandardScaler().fit_transform(X_test)
for k in range(3):
    X_test_cls = X_test[y_test == k]
    plt.plot(X_test_cls[:, 0], X_test_cls[:, 1], 'o')
```

Запустим EM-алгоритм на сгенерированных данных.

```python
model_test = em_gauss_mix_params(X_test, 3, verbose=True)
for k in range(3):
    X_test_cls = X_test[np.max(model_test['gamma'][:, :], axis=1) == model_test['gamma'][:, k]]
    plt.plot(X_test_cls[:, 0], X_test_cls[:, 1], 'o')
```

Можно видеть, что в результате EM-алгоритма была восстановлена смесь нормальных распределений из 3 компонент, и точкам были присвоены корректные кластеры.

Ниже приведен график значения логарифма правдоподобия в процессе итераций алгоритма.

```python
plt.figure(figsize=(10, 6))
plt.title('Log-likelihood over iterations')
plt.plot(np.arange(len(model_test['cost'])), model_test['cost'])
plt.xlabel('number of iterations')
plt.ylabel('log-likelihood')
```

Можно видеть, что логарифм правдоподобия не убывает с ходом итераций.

В верхнем ряду расположены изображения с максимальной ошибкой I рода, в нижнем ряду — с максимальной ошибкой II рода.

Для анализа качества кластеризации алгоритма вне зависимости от выбора порога отсечения построим ROC-кривую и вычислим значение AUC-ROC площади под кривой для разных значений параметра $\rho$ (ширины окна). Ниже построена данная кривая и вычислено значение AUC-ROC для неё. Кривая в данном случае строится приближённо, т.к. в зависимости от порога (threshold) в данном случае может меняться взаимная степень принадлежности пикселей классам.

```python
# Building AUC.
n_thrsh_opts = 30
n_rho_opts = 6

plt.figure(figsize=(8, 8))
plt.title('ROC curve')
plt.xlabel('FPR')
plt.ylabel('TPR')

fpr = np.empty(n_thrsh_opts)
tpr = np.empty(n_thrsh_opts)
thrsh_opts = np.linspace(0.0, 3.0, n_thrsh_opts)
rho_opts = np.linspace(0.0, 1.0, n_rho_opts)
for i, rho in enumerate(rho_opts):
    for j, t in enumerate(thrsh_opts):
        ped_res = gauss_1d_adaptive_predict(X_ped_gs, ped_gauss_1d['mu'], ped_gauss_1d['sigma'], t, rho)
        ped_errors = errors(y_ped, ped_res)
        ped_errors_sum = [ped_errors[i].sum() for i in range(4)]
        
        # fpr = fp / (fp + tn)
        # errors: (tp, fp, fn, tn)
        fpr[j] = ped_errors_sum[1] / (ped_errors_sum[1] + ped_errors_sum[3])
        # tpr = tp / (tp + fn)
        tpr[j] = ped_errors_sum[0] / (ped_errors_sum[0] + ped_errors_sum[2])
        #print('rho: {}, errors: {}, fpr: {}, tpr: {}'.format(rho, ped_errors_sum, fpr[j], tpr[j]))
    auc_roc = sklearn.metrics.auc(fpr, tpr, reorder=True)
    #TODO DataFrame with AUC stats
    print('AUC-ROC, rho={:.4}: {:.5}'.format(rho, auc_roc))
    plt.semilogy(fpr, tpr, label='ROC, rho={:.4}'.format(rho))

plt.legend()
```

Выбор небольшого значения параметра $\rho$ (ширины окна), например, около 0.2, немного повышает качество удаления фона по сравнению с неадаптивной моделью, хотя дает примерно то же качество.

Для анализа качества кластеризации алгоритма вне зависимости от выбора порога отсечения построим ROC-кривую и вычислим значение AUC-ROC площади под кривой. Ниже приближённо построена данная кривая и вычислено значение AUC-ROC для неё.

```python
tr_proba = em_predict_proba(X_tr, em_mu, em_sigma, em_w, tr_N, tr_D, tr_K)
idx_wb = (y_tr == 0) | (y_tr == 1)
fpr, tpr, _ = sklearn.metrics.roc_curve(y_tr[idx_wb], 1.0 - tr_proba[idx_wb])
roc_auc = sklearn.metrics.auc(fpr, tpr)

plt.figure(figsize=(8, 8))
plt.title('ROC curve')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.plot(fpr, tpr, label='AUC-ROC = {:.5}'.format(roc_auc))
plt.legend()
```

Сохраним получившиеся изображения в папку. Из полученных изображений можно затем составить видео.

```python
background_blending = 0.7

for n in range(tr_N):
    mask = tr_res[n].reshape(tr_shape[:2])
    img = X_tr[n].reshape(tr_shape).copy()
    #img[:, :, 1] = np.minimum(np.full_like(img[:, :, 1], 1.0), img[:, :, 1] + 100 * mask)
    img[~mask, :]  = background_blending * 1.0 + (1 - background_blending) * img[~mask, :]
    imlib.save_image('traffic_output/{:04}.png'.format(n), img)
```

#### Модификация предложенного метода (удаление шума и сглаживание границ)

В результате применения EM-алгоритма было получено разделение изображений на объект и фон. На изображениях присутствуют следующие ошибки:
* "шум" &mdash; мелкие точки на дороге принимаются за часть объекта
* небольшой "ореол" вокруг машины принимается за часть машины

(оба типа ошибок хорошо заметны на видео **em_video_1.mp4**)

Данные ошибки можно устранить, если считать пиксель принадлежащим объекту, только если большая часть его окрестности тоже принадлежит объекту. Это соответствует свертке с Uniform или Box фильтром (ядро свертки содержит одинаковые компоненты). Попробуем свернуть "маску" сегментации с Uniform фильтром.

```python
tr_sample = tr_res[95].reshape(tr_shape[:2])
tr_sample_uniform = scipy.ndimage.filters.uniform_filter(tr_sample, size=5)

fig, ax_arr = plt.subplots(1, 2, figsize=(13, 13))
ax_arr[0].set_title('Before (EM result)')
ax_arr[1].set_title('After (EM result + Uniform filter)')
ax_arr[0].imshow(tr_sample, cmap=matplotlib.cm.Greys_r)
ax_arr[1].imshow(tr_sample_uniform, cmap=matplotlib.cm.Greys_r)
ax_arr[0].axis('off')
ax_arr[1].axis('off')
print()
```

На изображении осталось несколько мелких точек. Попробуем использовать фильтр Гаусса вместе Uniform фильтра.

```python
tr_sample = tr_res[95].reshape(tr_shape[:2])
tr_sample_uniform = scipy.ndimage.filters.gaussian_filter(tr_sample, sigma=1)

fig, ax_arr = plt.subplots(1, 2, figsize=(13, 13))
ax_arr[0].set_title('Before (EM result)')
ax_arr[1].set_title('After (EM result + Gaussian filter)')
ax_arr[0].imshow(tr_sample, cmap=matplotlib.cm.Greys_r)
ax_arr[1].imshow(tr_sample_uniform, cmap=matplotlib.cm.Greys_r)
ax_arr[0].axis('off')
ax_arr[1].axis('off')
print()
```

Мелкий шум на изображении пропал. 

Преобразуем изображения, сделав свертку маски сегментации с фильтром Гаусса.

```python
tr_gauss_res = np.empty_like(tr_res)
for n in range(tr_N):
    cur_img = tr_res[n].reshape(tr_shape[:2])
    gauss = scipy.ndimage.filters.gaussian_filter(cur_img, sigma=1)
    tr_gauss_res[n] = gauss.reshape(tr_size)
```

Сохраним получившиеся изображения в папку. Из полученных изображений можно затем составить видео.

```python
background_blending = 0.7

for n in range(tr_N):
    mask = tr_gauss_res[n].reshape(tr_shape[:2])
    img = X_tr[n].reshape(tr_shape).copy()
    #img[:, :, 1] = np.minimum(np.full_like(img[:, :, 1], 1.0), img[:, :, 1] + 100 * mask)
    img[~mask, :]  = background_blending * 1.0 + (1 - background_blending) * img[~mask, :]
    imlib.save_image('traffic_output/{:04}.png'.format(n), img)
```

К отчету приложено видео **em_video_2.mp4**, составленное из полученного набора изображений (EM-алгоритм + фильтрация по Гауссу). На каждом кадре область объекта оставлена неизменной, область фона сделана прозрачной.
